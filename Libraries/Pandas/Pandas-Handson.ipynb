```python
# Access all the rows of a particular colums-

df['Hobbyist']
```

```python
# Instead of all rows access any column from the list of columns with specified rows using index

df.loc[[0,1],'Hobbyist']
```

```python
# Also we can paas multiple columns 

df.loc[[5,6],['ITperson','WorkChallenge']]
```

```python
# Check how many Yes/ No responses in the Hobbyist column

df['Hobbyist'].value_counts()
```

```python
# we can use slicing like a list to access multiple rows

df.loc[[0,1,2],['ITperson','Hobbyist','WorkChallenge']]

#df.loc[[0:2],['ITperson','Hobbyist','WorkChallenge']]   --> here we can't use braces if we using slicing then. use like this -

df.loc[0:2,['ITperson','Hobbyist','WorkChallenge']]
```

```python
# Slicing for columns -

df.loc[0:5, 'Hobbyist' : 'Country']
```

```python
# Similarly we can use slicing with iloc -

#0:2 -> row number 0 & 1
#1:6 -> column number 1,2,3,4,5    

df.iloc[0:2, 1:6]
```
